```

```

#Ejercicio 5
Crea un diccionario con nombres de estudiantes y sus notas.
Muestra el promedio de todos los estudiantes y de un estudiante específico, empleando dos funciones.

````python
def mostrar_nota_estudiante(estudiante):
    if estudiante in notas:
        print(f"Nota de {estudiante}:", notas[estudiante])
    else:
        print("Estudiante no encontrado.")

def mostrar_promedio():
    promedio = sum(notas.values()) / len(notas)
    print("Promedio del grupo:", promedio)

notas = {
    "Majo": 8.5,
    "Luis": 7.0,
    "Sofía": 9.2,
    "Carlos": 6.8
}

mostrar_nota_estudiante("Sofía")
mostrar_promedio()
````



```
import java.util.HashMap;

public class Main {
    static HashMap<String, Double> notas = new HashMap<>();

    static void mostrarNotaEstudiante(String estudiante) {
        if (notas.containsKey(estudiante)) {
            System.out.println("Nota de " + estudiante + ": " + notas.get(estudiante));
        } else {
            System.out.println("Estudiante no encontrado.");
        }
    }

    static void mostrarPromedio() {
        double suma = 0;
        for (double nota : notas.values()) {
            suma += nota;
        }
        System.out.println("Promedio del grupo: " + suma / notas.size());
    }

    public static void main(String[] args) {
        notas.put("Ana", 8.5);
        notas.put("Luis", 7.0);
        notas.put("Sofia", 9.2);
        notas.put("Carlos", 6.8);

        mostrarNotaEstudiante("Sofia");
        mostrarPromedio();
    }
}
```

#Ejercicio 7
Dada la palabra "magia", cuenta cuántas veces aparece cada letra usando un diccionario.

Pista:
Recorre la palabra letra por letra y ve acumulando el conteo.

````python
palabra = "magia"
conteo_letras = {}

for letra in palabra:
    if letra in conteo_letras:
        conteo_letras[letra] += 1
    else:
        conteo_letras[letra] = 1

print("Conteo de letras:", conteo_letras)
````



```
import java.util.HashMap;

public class Main {
    public static void main(String[] args) {
        String palabra = "magia";
        HashMap<Character, Integer> conteoLetras = new HashMap<>();

        for (char letra : palabra.toCharArray()) {
            if (conteoLetras.containsKey(letra)) {
                conteoLetras.put(letra, conteoLetras.get(letra) + 1);
            } else {
                conteoLetras.put(letra, 1);
            }
        }

        System.out.println("Conteo de letras: " + conteoLetras);
    }
}
```



#Ejercicio 8
Dada una lista de elementos recolectados en una expedición, cuenta cuántas veces aparece cada elemento.

Lista sugerida: ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]

````python
elementos = ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]
conteo = {}

for elemento in elementos:
    if elemento in conteo:
        conteo[elemento] += 1
    else:
        conteo[elemento] = 1

print("Conteo de elementos:", conteo)
````



```
import java.util.HashMap;

public class Main {
    public static void main(String[] args) {
        String[] elementos = {"gema", "oro", "gema", "madera", "oro", "oro", "gema"};
        HashMap<String, Integer> conteo = new HashMap<>();

        for (String elemento : elementos) {
            if (conteo.containsKey(elemento)) {
                conteo.put(elemento, conteo.get(elemento) + 1);
            } else {
                conteo.put(elemento, 1);
            }
        }

        System.out.println("Conteo de elementos: " + conteo);
    }
}
```



#Pregunta (2000exp): Qué otras funciones poseen los diccionarios?

-RTA:
Los diccionarios en Python tienen varias funciones que sirven para trabajar mejor con los datos. Por ejemplo, keys muestra todas las claves, values muestra los valores y items muestra ambos (clave y valor). También está get, que sirve para obtener un valor sin que dé error si la clave no existe.

Además, update permite agregar o cambiar datos, pop elimina un elemento, y clear borra todo el diccionario. La función copy hace una copia, fromkeys crea un diccionario nuevo con claves dadas, y setdefault agrega una clave si no existe.
