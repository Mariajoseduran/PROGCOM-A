

```
import java.util.Scanner;

public class Ej18 {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int factorial = 1;
        int i = 1;

        if(n < 0){
            System.out.println("No hay factorial negativo");
        }else{
            while(i <= n){
                factorial *= i;
                i++;
            }
            System.out.println("Factorial: " + factorial);
        }
    }
}```



ejercicio 19

````python
n = int(input("¿Cuántos números de Fibonacci desea?: "))

a = 0
b = 1
i = 0

while i < n:
    print(a)
    temp = a + b
    a = b
    b = temp
    i = i + 1
````

ejercicio 20

````python
colores = []
i = 0

while i < 10:
    color = input("Ingrese un color: ")
    colores.append(color)
    i = i + 1

secreto = colores[0]
intentos = 0

while intentos < 10:
    intento = input("Adivine el color: ")

    if intento == secreto:
        print("Correcto")
        break
    else:
        print("Incorrecto")

    intentos = intentos + 1
````

ejercicio 22

````python
numero = int(input("Ingrese un número: "))
suma = 0

while numero > 0:
    digito = numero % 10
    suma = suma + digito
    numero = numero // 10

print("La suma de los dígitos es:", suma)
````

ejercicio 23

````python
numero = int(input("Ingrese un número: "))
i = 2
primo = True

while i < numero:
    if numero % i == 0:
        primo = False
    i = i + 1

if primo and numero > 1:
    print("Es primo")
else:
    print("No es primo")
````
