````python
#quiero sabr cuantos de n estudiantes sacan mas que el promedio
suma=0
c=0
estudiantes= int(input("Cuantos estudiantes son?"))
for i in range(estudiantes):
  notas=float(input("Ingrese las notas de cada uno:"))
  suma+=notas
promedio=suma/estudiantes
print(promedio)
for i in range(estudiantes):
  if notas>promedio:
    c=c+1
print(f"{c}estudiantes sacaron mayor nota que el promedio")
````
