Tabla de multiplicar de un número hasta el 10 python

````python
numero = int(input("Ingrese un número: "))

for i in range(1, 11):
    print(numero, "x", i, "=", numero * i)
````

````python
numero = int(input("Ingrese un número: "))
i = 1

while i <= 10:
    print(numero, "x", i, "=", numero * i)
    i += 1
````

import java.util.Scanner;

public class TablaFor {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        
        System.out.print("Ingrese un número: ");
        int numero = sc.nextInt();

        for (int i = 1; i <= 10; i++) {
            System.out.println(numero + " x " + i + " = " + (numero * i));
        }

        sc.close();
    }
}```

import java.util.Scanner;

public class TablaWhile {

        System.out.print("Ingrese un número: ");

        int i = 1;

        }

Suma de números naturales hasta n

````python
n = int(input("Ingrese un número: "))
suma = 0

for i in range(1, n + 1):
    suma += i

print("La suma es:", suma)
````

````python
n = int(input("Ingrese un número: "))
suma = 0
i = 1

while i <= n:
    suma += i
    i += 1

print("La suma es:", suma)
````

Factorial de un número

````python
n = int(input("Ingrese un número: "))

if n < 0:
    print("No existen factoriales negativos")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    print("El factorial es:", factorial)
````

````python
n = int(input("Ingrese un número: "))

if n < 0:
    print("No existen factoriales negativos")
else:
    factorial = 1
    i = 1
    while i <= n:
        factorial *= i
        i += 1
    print("El factorial es:", factorial)
````

Serie Fibonacci (primeros n números)

````python
n = int(input("¿Cuántos números desea ver?: "))
a, b = 0, 1

for i in range(n):
    print(a)
    a, b = b, a + b
````

````python
n = int(input("¿Cuántos números desea ver?: "))
a, b = 0, 1
i = 0

while i < n:
    print(a)
    a, b = b, a + b
    i += 1
````

Adivinar un color (máximo 10 intentos)

````python
colores = []
for i in range(10):
    colores.append(input("Ingrese un color: "))

import random
color_secreto = random.choice(colores)

intentos = 0
while intentos < 10:
    intento = input("Adivine el color: ")
    intentos += 1

    if intento == color_secreto:
        print("¡Correcto!")
        break
    else:
        print("Incorrecto")

if intentos == 10:
    print("Se acabaron los intentos. El color era:", color_secreto)
````

Suma de los dígitos de un número

````python
numero = int(input("Ingrese un número: "))
suma = 0

while numero > 0:
    suma += numero % 10
    numero //= 10

print("La suma de los dígitos es:", suma)
````

Determinar si un número es primo

````python
numero = int(input("Ingrese un número: "))
es_primo = True

if numero <= 1:
    es_primo = False
else:
    for i in range(2, numero):
        if numero % i == 0:
            es_primo = False
            break

if es_primo:
    print("Es primo")
else:
    print("No es primo")
````
