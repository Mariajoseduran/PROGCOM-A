#EJERCICIO 1

````python
nombres = ["Ana", "Juan", "Pedro", "Maria", "Luis", "Elena", "Pablo", "Lucia", "Jose", "Rosa"]
notas = [85, 90, 70, 95, 80, 75, 100, 88, 60, 92]

promedio = sum(notas) / 10
print("El promedio de la clase es:", promedio)

print("Estudiantes por encima del promedio ")

for i in range(10):
    if notas[i] > promedio:
        print(nombres[i], "sacó", notas[i])

nota_mayor = max(notas)
posicion = notas.index(nota_mayor)
nombre_mayor = nombres[posicion]

print("El mejor fue", nombre_mayor, "con una nota de:", nota_mayor)
````

##EJERCICIO 2

````python
n = int(input("Ingrese el número de votantes: "))

candidatos = ["abelardo", "cepeda", "paloma", "vicky", "botero"]
votos = [0, 0, 0, 0, 0]

print("\n--- Primera vuelta ---")
print("Opciones:")
for i in range(5):
    print(i+1, "-", candidatos[i])

for i in range(n):
    voto = int(input(f"Votante {i+1}, elija un candidato (1-5): "))
    if 1 <= voto <= 5:
        votos[voto-1] += 1
    else:
        print("Voto inválido")

print("\nResultados de la primera vuelta:")
for i in range(5):
    print(candidatos[i], ":", votos[i], "votos")

max1 = 0
max2 = 1

if votos[max2] > votos[max1]:
    max1, max2 = max2, max1

for i in range(2, 5):
    if votos[i] > votos[max1]:
        max2 = max1
        max1 = i
    elif votos[i] > votos[max2]:
        max2 = i

print("\nPasan a segunda vuelta:")
print(candidatos[max1])
print(candidatos[max2])

votos2 = [0, 0]

print("\n--- Segunda vuelta ---")
print("1 -", candidatos[max1])
print("2 -", candidatos[max2])

for i in range(n):
    voto = int(input(f"Votante {i+1}, elija (1-2): "))
    if voto == 1:
        votos2[0] += 1
    elif voto == 2:
        votos2[1] += 1
    else:
        print("Voto inválido")

print("\nResultados de la segunda vuelta:")
print(candidatos[max1], ":", votos2[0], "votos")
print(candidatos[max2], ":", votos2[1], "votos")

if votos2[0] > votos2[1]:
    print("\nGanador:", candidatos[max1])
elif votos2[1] > votos2[0]:
    print("\nGanador:", candidatos[max2])
else:
    print("\nEmpate en la segunda vuelta")
````

##EJERCICIO 3

````python
print("MENU")
menu = {
    "1": ("Combo 1",        20000),
    "2": ("Combo 2",        25000),
    "3": ("Combo 3",        18000),
    "4": ("Combo 4",        24500),
    "5": ("Papas",           3000),
    "6": ("Jugos naturales", 7000),
    "7": ("Cerveza",         3500),
    "8": ("Agua",            4000),
    "9": ("Gaseosas",        2500),
}

total_dia = 0
ventas = {v[0]: 0 for v in menu.values()}

print("\nMENÚ:")
for k, (nombre, precio) in menu.items():
    print(f"  {k}. {nombre}: ${precio:,}")
print("  Escriba SALIR para cerrar caja")

continuar = 1

while continuar == 1:
    opcion = input("\nORDEN DEL CLIENTE O ESCRIBA SALIR: ")

    if opcion == "SALIR":
        continuar = 0
    elif opcion in menu:
        nombre, precio = menu[opcion]
        total_dia += precio
        ventas[nombre] += 1
        print(f"{nombre} Total acumulado: ${total_dia:,}")
    else:
        print("Opción no válida")

print("\nRESUMEN DEL DÍA ")
for producto, cantidad in ventas.items():
    if cantidad > 0:
        precio = [v[1] for v in menu.values() if v[0] == producto][0]
        print(f"{producto}: {cantidad} unidades = ${cantidad * precio:,}")

print(f"\nTOTAL: ${total_dia:,}")
````
