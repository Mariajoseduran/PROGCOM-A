PARA- FOR
 * Se emplea cuando se sabe cuantas veces se repetira la acción
 **OJO**  El conteo siempre inicia desde **0**

 sintaxis general


```
for (variable) in (algo):
 la acción que se va a repetir

 #forma 2
 for variable in range (VI,VF+1,P):
 La accion que se va a repetir
```
donde:
* VI: Valor inicial (dónde inicio)
* VF: valor final (dónde finalizó)
* P: Paso. cada cuánto cuento




sintaxis para en java


```
for (int variable=VI;variable=VF;variable=variable+paso){La accion que se va a repetir.}
```



````python
#Ejercicio: imprimir los numeros pares del 1 al 10
for num in range(11):
 if num%2==0: print(num)
#poner un numero mas del que quiero en el rango ya que cuenta desde el 0
````
