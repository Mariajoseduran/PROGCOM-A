 programa que calcule el índice de masa corporal (IMC) y determine la categoría

````python
peso=float(input("Ingrese su peso kg: "))
altura=float(input("Ingrese su estatura m: "))
imc=peso/(altura**2)
print(f"Este es su IMC {imc:.2f}")
if imc<18.5:
  print("Tiene bajo peso")
elif 18.5<imc<24.9:
  print("Tiene peso normal")
elif 25<imc<29.9:
  print("Tiene sobrepeso")
else:
  print("Tiene obesidad")
````



```
import java.util.*;
public class Main
{
    public static void main(String[] args){
        Scanner entrada = new Scanner(System.in);

       // 1. Entrada de datos
        System.out.print("Ingrese su peso kg: ");
        double peso = entrada.nextDouble();

        System.out.print("Ingrese su estatura m: ");
        double altura = entrada.nextDouble();

        // 2. Cálculo del IMC
        double imc = peso / (altura * altura);

        // Mostrar el IMC con 2 decimales
        System.out.printf("Este es su IMC: %.2f%n", imc);

        // 3. Clasificación con if - else if - else
        if (imc < 18.5) {
            System.out.println("Tiene bajo peso");
        }
        else if (imc >= 18.5 && imc <= 24.9) {
            System.out.println("Tiene peso normal");
        }
        else if (imc >= 25 && imc <= 29.9) {
            System.out.println("Tiene sobrepeso");
        }
        else {
            System.out.println("Tiene obesidad");
        }

        entrada.close();
    }
}
```

```
import java.util.*;
public class Main
{
    public static void main(String[] args){
        Scanner entrada = new Scanner(System.in);

        // 1. Solicitar datos (usamos double para mayor precisión en cálculos)
        System.out.print("Ingrese su peso en kg: ");
        double peso = entrada.nextDouble();

        System.out.print("Ingrese su estatura en m: ");
        double altura = entrada.nextDouble();

        // 2. Calcular IMC: peso / (altura * altura)
        double imc = peso / (altura * altura);

        // Mostrar el IMC con formato de dos decimales
        System.out.printf("Este es su IMC: %.2f%n", imc);

        // 3. Determinar categoría usando SI TERNARIO anidado
        String resp = (imc < 18.5)
            ? "Tiene bajo peso"
            : (imc <= 24.9)
                ? "Tiene peso normal"
                : (imc <= 29.9)
                    ? "Tiene sobrepeso"
                    : "Tiene obesidad";

        System.out.println(resp);

        entrada.close();
    }
}
```
