PARA- FOR
 * Se emplea cuando se sabe cuantas veces se repetira la acción
 **OJO**  El conteo siempre inicia desde **0**

 sintaxis general


```
for (variable) in (algo):
 la acción que se va a repetir

 #forma 2
 for variable in range (VI,VF+1,P):
 La accion que se va a repetir
```
donde:
* VI: Valor inicial (dónde inicio)
* VF: valor final (dónde finalizó)
* P: Paso. cada cuánto cuento




sintaxis para en java


```
for (int variable=VI;variable=VF;variable=variable+paso){La accion que se va a repetir.}
```



````python
#Ejercicio: imprimir los numeros pares del 1 al 10
for num in range(11):
 if num%2==0: print(num)
#poner un numero mas del que quiero en el rango ya que cuenta desde el 0
````

````python
import random

jugando = True

while jugando:
    eleccion = input("Elige entre 1. Piedra, 2. Papel, 3. Tijera: ")

    while eleccion != "1" and eleccion !="2" and eleccion !="3":
      eleccion=input("Valor no valido")

    yo = int(eleccion)
    m = random.randint(1, 3)
    print(f"La máquina eligió: {m}")

    if yo == m:
        print("empate. Ambos elegimos 🪨")
    elif yo == 1 and m == 2:
        print("Perdiste, La maquina eligio 2")
    elif yo == 2 and m == 3:
        print("Perdiste, La maquina eligio")
    elif yo == 3 and m == 1:
        print("Perdiste, La maquina eligio")
    else:
        print("Ganaste")

    continuar = input("¿Quieres seguir jugando? (si/no): ")
    if continuar == "no":
        jugando = False
        print("Gracias por jugar. ¡Hasta la próxima!")
````
